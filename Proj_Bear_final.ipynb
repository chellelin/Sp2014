##### Data cleaning & reformatting

```python
%cd "/Users/Cheng/Documents/Spring2014/Stat222/bear"

bear = pd.read_csv('bear.csv')
bear['Call Date'] = pd.to_datetime(bear['Call Date']) # convert date into a better format
### several waiting times are recorded as 23:59, which is unrealistic
bear = bear[bear['Student wait time']!='23:59']
bear = bear[['Call Date','Hour','Student wait time','No. of Passengers']].dropna() # get rid of missing data
```

```python
# define a function to convert waiting time into integers
def convMin(x):
    (h, m) = x.split(':')
    result = int(h) * 60 + int(m) 
    return result
vec_convMin = np.vectorize(convMin) 
```

```python
bear['Student wait time'] = vec_convMin(bear['Student wait time'].values) # convert waiting time into integers
bear['waiting time sum'] = bear['No. of Passengers']*bear['Student wait time'] # add a column of "waiting time sum"
bear[:3]
```

```python
# get a new data frame with a column of average waiting time in each hour
new_bear = bear.groupby(['Call Date','Hour']).sum()[['No. of Passengers','waiting time sum']]
new_bear["avg waiting time"]  = new_bear["waiting time sum"]/new_bear["No. of Passengers"] 
new_bear[:10]
```

```python
# get a series of waiting time on a 24 hour base, fill waiting time in the hours with no request as zero
wt = new_bear['avg waiting time']
wt = wt.unstack().fillna(0)
wt = wt.reindex(columns=range(0,24)).fillna(0)
wt = wt.stack()
wt_Fall = wt[432:3288] # limit the series on the time interval 8/19 - 12/15
wt_Fall.index.names = ['Call Date', 'Hour']
# wt_Fall.head()
```

```python
# do the same for no. of passengers
pt = new_bear['No. of Passengers']
pt = pt.unstack().fillna(0)
pt = pt.reindex(columns=range(0,24)).fillna(0)
pt = pt.stack()
pt_Fall = pt[432:3288]
pt_Fall.index.names = ['Call Date', 'Hour']
```

##### 2.2.2 ACF & Seasonality

  &nbsp;&nbsp;&nbsp;&nbsp;Why doesn't autogression fit? Our preliminary analysis on the daily pattern of waiting time may give us the answer. To confirm this 24 hour pattern. Let's look at the ACF and PACF of the time series. Indeed, ACF gives further support for a seasonal model as shown in the Figure below. The oscillation of ACF suggests seasonality with a period of 1 day, or 24 hours. Hence "failure to take seasonality into account" is the main reason of bad fit of the autogression model above. 

```python
# ACF and PACF analysis
fig = plt.figure(figsize=(12,8))
# change ticks here ticks = ax.set_xticks(range(0,300,24))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(ts.values.squeeze(), lags=300, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(ts, lags=300, ax=ax2)
```

   &nbsp;&nbsp;&nbsp;&nbsp;From ACF，it seems that there is a seasonal pattern of 24 hours in the waiting time. Let's take the seasonal difference with a lag of 24 hours, and see how the series of seasonal difference perform:

```python
sd_ts = ts[:-24].values - ts[24:].values
sd_ts = pd.Series(sd_ts,index = hours[:-24])
sd_ts.plot()
# ACF and PACF analysis
fig = plt.figure(figsize=(12,8))
# change ticks here ticks = ax.set_xticks(range(0,300,24))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(sd_ts.values.squeeze(), lags=300, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(sd_ts, lags=300, ax=ax2)
```

  &nbsp;&nbsp;&nbsp;&nbsp;The seasonal difference series looks stable. ACF plot of the seasonal differenced series suggests a seasonal MA(1) and the exponentially decaying PACF further supports it being an MA process. Hence I will set down (P,D,Q) = (0,0,1)

##### 2.2.3 A better solution: SARIMAX Model

  &nbsp;&nbsp;&nbsp;&nbsp;Now I'd like to use a multiplicative seasonal ARIMA model, since it is unlikely that the true model incorporate auto correlation only at the seasonal lags. However, It seems to me that Python's statsmodels module can only handle SARIMA model. So I will use rmagic to call some R code here.

```python
%pylab inline
%load_ext rmagic 
```

  &nbsp;&nbsp;&nbsp;&nbsp;My first trials are Seasonal ARIMA and Seasonal Means Model. However, ACF of residuals rejects their independence in both models. A cosine and sine linear fit based on periodogram gives similar results.  
  &nbsp;&nbsp;&nbsp;&nbsp;After the failure of fitting the data in SARIMA and seasonal means model, I decide to combine the two together. I use the hourly expected average waiting time as explanatory variables added to a SARIMA (p,d,q) ×(P,D,Q) model. The process is represented as follows:

<h6 align="center"> $\varphi_p(B)\Phi_P(B_s)W_t = \theta_q(B)\Theta_Q(B_s)e_t + \alpha \mu_t$</h6>
<h6 align="center">$W_t = \nabla^d\nabla^DY_t$</h6>
<h6 align="center">$\mu_t = \beta_i $ where i = t (mod 24)</h6> 

  &nbsp;&nbsp;&nbsp;&nbsp; B represents the backwards shift operator, and $e_t$ is a white noise process. $W_t$ denotes the differenced series. $\varphi_p$, $\Phi_P$, $\theta_q$, $\Theta_Q$ are polynomials of order p, P, q, Q, each determined by its corresponding autoregressive or moving average terms. The 24 constants, β0,β1,... and β23 are the expected average waiting time of each hour given by Seasonal Means Model.

```python
test = ts.values
```

```python
%%R -i test
wt = test
wt = ts(wt,frequency=24,start=c(0,1))
```

&nbsp;&nbsp;&nbsp;&nbsp;We already know that (P,D,Q) = (0,0,1) from the section above. Set xreg =$\mu_t$, and $\beta_i$'s are calculated using the code below from seasonal mean.   

```python
%%R
# seasonal means => get beta's
library("TSA");
hour.=season(wt)
wt.m3=lm(wt~hour.-1)
# summary(wt.m3) #R^2 = 0.7525
# AIC(wt.m3,k=log(n)) #17803.62 # 9676.299
# acf(residuals(wt.m3),lag.max=200) # :(
```

&nbsp;&nbsp;&nbsp;&nbsp;Then I calculate BIC of various conbinations of (p,0,q) as an attempt to determine the optimal process. The BIC plot reaches its minimum value 18232.65 at (p,q) = (1,1), suggesting a SARIMA(1,0,1) ×(0,0,1). The BIC is significantly better than that of either pure SARIMA or seasonal means model. 

```python
%%R -i n
n = n
#xreg seasonal mean
sa<-wt.m3[[1]] # seasonal mean
hr<-rep(sa,119)

# this takes a long time
sbic<-rep(NA,25)
pp<-rep(NA,25)
qq<-rep(NA,25)
i=1
for(p in 0:4){
  for(q in 0:4){
    pp[i]<-p
    qq[i]<-q
    sbic[i]=AIC(arima(wt,order=c(p,0,q),seasonal=list(order=c(0,0,1),period=24),xreg=hr),k=log(n))
    i=i+1
  }
}
```

##### 2.2.4: Other Candidates

&nbsp;&nbsp;&nbsp;&nbsp; Several other models also have performance comparable to that of SARIMAX. However, I find SARIMAX the easiest one to apply. For example, spurious correlation model returns that waiting time is led by number of requests with lag 0 and lag 1. Even though the result makes intuitive sense, the model is not practical because number of requests also requires prediction. Moreover, it is unlikely to be a one-way influence. Waiting time may in turn affects number of requests, so the model based on spurious correlation is not reliable. Another one I have looked into is STL decompositoin. Basically the model decomposes the time series in to three parts, short term daily trend, long term semester trend and errors. This is not a practical model either because the long term pattern is not externally valid. Fall semester is very different from Spring semester, and the pattern may change from year to year. 

<h2 align = "center">3. Conclusion </h2>

<p>&nbsp;&nbsp;&nbsp;&nbsp; This Report attempts to capture and gain some insight in to BearWalk waiting time pattern using various seasonal models. BearWalk waiting time has obvious seasonal pattern with lag of 24 hours. However, the current prediction scheme used by BearWalk does not take this seasonality into account. While pure SARIMA model does not represent the series well, SARIMAX model using expected average waiting time as explanatory variables improves the fit significantly. Spurious correlation and STL decomposition models both have their strength in fitting waiting time, but they are not as useful as SARIMAX for practical perpose.</p>
<p>&nbsp;&nbsp;&nbsp;&nbsp; There are many hidden variables for BearWalk that I have not been taken into account in this analysis, such as special regular users like Security Monitors, weather condition, school calendar.To better predict waiting time, I suggests further study in SARIMAX with more explanatory varibles mentioned above. Of course the problem can also be studied at a totally different angle, such as prediction based on location. However, this would require extra investment in monitoring devices and prediction based on times series is apparently more economical. </p>
